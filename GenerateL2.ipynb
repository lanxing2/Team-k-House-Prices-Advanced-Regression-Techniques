```python
# load data
import pandas as pd
import numpy as np
from pandas import DataFrame
base_path='D:/kaggle/regression/'
all_data=DataFrame.from_csv(base_path+'paragon_train.csv')
all_data.info()
```

```python
all_id=all_data.index
all_y=all_data['SalePrice']
print all_y

all_x=all_data.drop(['SalePrice'],axis=1)
all_x.describe()
```

```python
# NAN and Inf check
def validate_df(df):
    # Print columns that contains NAN
    print 'Checking NAN'
    for k,v in df.isnull().any().iteritems():
        if v:
            print 'Column ',k,' contains NAN'
    
    # Print columns that contains Infinity
    print 'Infinity values in dataframe: ',
    print df[np.isinf(df)].stack().dropna()

def validate_series(s):
    print 'Series contains NAN? ',
    print pd.isnull(s).any()
    
    print 'Inf value in series: ',
    print s[np.isinf(s)].dropna()
```

```python
validate_df(all_x)

validate_series(all_y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(all_x, all_y, test_size=0.2, random_state=42)
```

```python
from sklearn.metrics import mean_squared_error

def rmse_log_error(predict,reality):
    predict_log=np.log(predict)
    reality_log=np.log(reality)
    
    # If the prediction is no greater than 0, log will give NAN or INF
    predict_log=np.nan_to_num(predict_log)
    reality_log=np.nan_to_num(reality_log)
    
    mse=mean_squared_error(reality_log, predict_log)
    rmse=np.sqrt(mse)
    return rmse
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR
import xgboost as xgb
from sklearn.linear_model import Lasso

# A function to train several models for the train/test pair for one fold
# How to tune parameters is out of scope of this script, check out sklearn's GridSearchCV or RandomizedSearchCV
def one_fold(train_x,train_y,test_x,test_y, full_X, full_y):
    # Report shape
    print "Training set has shape: ",X_train.shape
    print "Test set has shape: ",X_test.shape
    print 'Train the model based on all known data ',full_X.shape
    
#     # Lasso
#     print 'Lasso'
#     las = Lasso(alpha=0.00099, max_iter=50000)
#     print 'Fitting Lasso'
#     las.fit(train_x,train_y)
#     print 'Predicting on test set'
#     las_result=las.predict(test_x)
#     print las_result
#     print('Lasso RMSE {score}'.format(score=rmse_log_error(test_y, las_result)))
    
#     las.fit(full_X,full_y)

    # Random forest with mse, auto-tuned
    print "Random forest with mse"
    rf0 = RandomForestRegressor(criterion='mse', n_estimators=126,max_features=0.5,max_depth=10,random_state=42)
    print "Fitting random forest with mse"
    rf0.fit(train_x, train_y)
    print 'Predicting on test set'
    rf0_result=rf0.predict(test_x)
    print('RF Auto RMSE {score}'.format(score=rmse_log_error(test_y, rf0_result)))
    
    rf0.fit(full_X,full_y)
    
    # Random forest with mse, hand-tuned
    print "Random forest with mse"
    rf1 = RandomForestRegressor(criterion='mse', n_estimators=250,max_features=0.21,max_depth=31,random_state=42)
    print "Fitting random forest with mse"
    rf1.fit(train_x, train_y)
    print 'Predicting on test set'
    rf1_result=rf1.predict(test_x)
    print('RF Manual RMSE {score}'.format(score=rmse_log_error(test_y, rf1_result)))
    
    rf1.fit(full_X,full_y)
    
    # Extra tree regressor, auto-tuned
    print "ExtraTreesRegressor, auto-tuned"
    et0 = ExtraTreesRegressor(criterion='mse', n_estimators=180,max_features=0.2,max_depth=28,n_jobs=-1,random_state=42)
    print "Fitting extra trees regressor with mse"
    et0.fit(train_x,train_y)
    print 'Predicting on test set'
    et0_result=et0.predict(test_x)   
    print('ET RMSE {score}'.format(score=rmse_log_error(test_y, et0_result)))
    
    et0.fit(full_X,full_y)
    
    # Extra tree regressor, hand-tuned
    print "ExtraTreesRegressor, hand-tuned"
    et1 = ExtraTreesRegressor(criterion='mse', n_estimators=600,max_features=0.475,
                             min_samples_split=4,min_samples_leaf=2,
                             max_depth=16,n_jobs=-1,random_state=42)
    print "Fitting extra trees regressor with mse"
    et1.fit(train_x,train_y)
    print 'Predicting on test set'
    et1_result=et1.predict(test_x)   
    print('ET RMSE {score}'.format(score=rmse_log_error(test_y, et1_result)))
    
    et1.fit(full_X,full_y)
    
    # Adaboost linear
    print 'Adaboost linear'
    ad0=AdaBoostRegressor(loss='linear',learning_rate=0.2,n_estimators=210,random_state=42)
    print 'Fitting adaboost linear'
    ad0.fit(train_x,train_y)
    print 'Predicting on test set'
    ad_result=ad0.predict(test_x)  
    print('Adaboost Linear RMSE {score}'.format(score=rmse_log_error(test_y, ad_result)))
    
    ad0.fit(full_X,full_y)
    
    # Adaboost square
    print 'Adaboost square'
    ad1=AdaBoostRegressor(loss='square',learning_rate=0.526,n_estimators=410,random_state=42)
    print 'Fitting adaboost linear'
    ad1.fit(train_x,train_y)
    print 'Predicting on test set'
    ad1_result=ad1.predict(test_x)  
    print('Adaboost Square RMSE {score}'.format(score=rmse_log_error(test_y, ad1_result)))
    
    ad1.fit(full_X,full_y)
    
    # Adaboost exp
    print 'Adaboost exponential'
    ad2=AdaBoostRegressor(loss='exponential',learning_rate=0.342,n_estimators=158,random_state=42)
    print 'Fitting adaboost linear'
    ad2.fit(train_x,train_y)
    print 'Predicting on test set'
    ad2_result=ad2.predict(test_x)  
    print('Adaboost Exponential RMSE {score}'.format(score=rmse_log_error(test_y, ad2_result)))
    
    ad2.fit(full_X,full_y)
    
    # GBM
    print 'Gradient Boosting Regressor'
    gbr=GradientBoostingRegressor(learning_rate=0.05,min_samples_split=110,max_depth=5,
                              min_samples_leaf=23,max_features=0.211,subsample=0.81,
                              random_state=42)
    print 'Fitting GBM'
    gbr.fit(train_x,train_y)
    print 'Predicting on test set'
    gbr_result=gbr.predict(test_x)  
    print('GBM RMSE {score}'.format(score=rmse_log_error(test_y, gbr_result)))
    
    gbr.fit(full_X,full_y)
    
    # XGBoost
    print 'XgBoost regressor'
    xgbr = xgb.XGBRegressor(
                 colsample_bytree=0.2,
                 gamma=0.0,
                 learning_rate=0.01,
                 max_depth=4,
                 min_child_weight=1.5,
                 n_estimators=7200,                                                                  
                 reg_alpha=0.9,
                 reg_lambda=0.6,
                 subsample=0.2,
                 seed=42,
                 silent=1)
    print 'Fitting XGB'
    xgbr.fit(train_x,train_y)
    print 'Predicting on test set'
    xgbr_result=xgbr.predict(test_x)
    print('XGB RMSE {score}'.format(score=rmse_log_error(test_y, xgbr_result)))
    
    xgbr.fit(full_X,full_y)
    
    # SVM rbf
    print 'SVM with rbf'
    svr0=SVR(kernel='rbf',coef0=0.0,shrinking=True,verbose=False,C=77,
             tol=0.001,epsilon=0.1,gamma=0.01)
    print 'Fitting SVR'
    svr0.fit(train_x,train_y)
    print 'Predicting on test set'
    svr0_result=svr0.predict(test_x)  
    print('SVR rbf RMSE {score}'.format(score=rmse_log_error(test_y, svr0_result)))
    
    svr0.fit(full_X,full_y)
    
    # SVM sigmoid
    print 'SVM with sigmoid'
    svr1=SVR(kernel='sigmoid',coef0=0.0,shrinking=True,verbose=False,C=77,
             tol=0.001,epsilon=0.1,gamma='auto')
    print 'Fitting SVR'
    svr1.fit(train_x,train_y)
    print 'Predicting on test set'
    svr1_result=svr1.predict(test_x)  
    print('SVR sigmoid RMSE {score}'.format(score=rmse_log_error(test_y, svr1_result)))
    
    svr1.fit(full_X,full_y)
    
    regressors={
        'rf0':rf0,
        'rf1':rf1,
        'et0':et0,
        'et1':et1,
        'ad0':ad0,
        'ad1':ad1,
        'ad2':ad2,
        'gbr':gbr,
        'svr0':svr0,
        'svr1':svr1,
        'xgbr':xgbr
#         'las':las
    }
    
    return regressors
```

```python
from sklearn.model_selection import train_test_split

regressors={}
for i in range(10):
    print 'Iteration ',i
    X_train, X_test, y_train, y_test=train_test_split(all_x, all_y, test_size=0.2, random_state=i)
    
    # Validate
    print 'Validate data first:'
    validate_df(X_train)
    validate_series(y_train)
    validate_df(X_test)
    validate_series(y_test)
    validate_df(all_x)
    validate_series(all_y)
    
    # Start
    print '\nStart training'
    batch=one_fold(X_train, y_train, X_test, y_test,all_x,all_y)
    regressors['rf_auto_'+str(i)]=batch['rf0']
    regressors['rf_manual_'+str(i)]=batch['rf1']
    regressors['et_auto_'+str(i)]=batch['et0']
    regressors['et_manual_'+str(i)]=batch['et1']
    regressors['ad_linear_'+str(i)]=batch['ad0']
    regressors['ad_sqr_'+str(i)]=batch['ad1']
    regressors['ad_exp_'+str(i)]=batch['ad2']
    regressors['gbr_'+str(i)]=batch['gbr']
    regressors['svr_rbf_'+str(i)]=batch['svr0']
    regressors['svr_sigmoid_'+str(i)]=batch['svr1']
    regressors['xgbr_'+str(i)]=batch['xgbr']
#     regressors['las_'+str(i)]=batch['las']
    
    
print len(regressors),' regressors trained'
```

```python
# The predictions then form the training samples for an L2 model as an ensemble.
l2_data=pd.DataFrame()
for i in range(10):
    reg_names=['rf_auto_'+str(i),'rf_manual_'+str(i),
               'ad_linear_'+str(i),'ad_sqr_'+str(i),'ad_exp_'+str(i),
               'gbr_'+str(i),
               'et_auto_'+str(i),'et_manual_'+str(i),
               'svr_rbf_'+str(i),'svr_sigmoid_'+str(i),
               'xgbr_'+str(i)
              ] # find regressors by name    
    
    print "This batch contains: ",reg_names
    for reg_name in reg_names:
        reg=regressors[reg_name]
            
        # Generate L2 on all data
        l2_pred=reg.predict(all_x)
        l2_data[reg_name]=l2_pred
        
    print "End of iteration ",i
    
print 'All regressors have predicted'
l2_data.index=all_id
l2_data.describe()
```

```python
# Append y to l2 data
print l2_data.shape
print len(all_y)
l2_full=pd.concat([l2_data,all_y],axis=1)
l2_full.index=all_id
l2_full['SalePrice'].describe()
```

```python
# Output to csv
l2_full.to_csv(base_path+'l2_boosting_train.csv',index_label='Id')
```
